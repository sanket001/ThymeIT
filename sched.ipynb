```python
import numpy as np
import csv
import collections
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_s = pd.read_csv('Student Data.csv')
df_s = pd.DataFrame(data_s, columns = ['Student ID', 'Student Name', 'Batch', 'Course Code'])
```

```python
#creating a dictionary for course group
cg = {}

#creating a function to add to the dictionary 

def course_group(file_obj):
    data = csv.DictReader(file_obj, delimiter=',')
    ccode = ""
        
    for line in data:
        val =  cg.get(line["Course Code"],"")
        if ( val == ""):
            ccode = line["Course Code"]
            ls = []
            ls.append(line["Student ID"])
            cg[ccode] = ls
        else: 
            val.append(line["Student ID"])
    
    print(cg)
#----- end of function

#calling the function

with open("Student Data.csv") as f_obj:
       course_group(f_obj)
```
